# 3D Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator
```

```python
# Make data.
x = np.arange(-4, 8, 0.25)
y = np.arange(-4, 2, 0.25)
x, y = np.meshgrid(x, y)
# R = np.sqrt(X**2 + Y**2)
# Z = np.sin(R)
z = 3*y**2 + 2*y**3 - 3*x**2 - 6*x*y
# 3y^2 + 2y^3 - 3x^2 - 6xy
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
```

```python
surf = ax.plot_surface(x, y, z, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8,
                alpha=0.3)
```

```python
# Customize the z axis.
ax.set_zlim(-100, 50)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.view_init(elev=30, azim=-60, roll=0)
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# ax.contour(X, Y, Z, zdir='z', offset=-100, cmap='coolwarm')
# ax.contour(X, Y, Z, zdir='x', offset=-40, cmap='coolwarm')
# ax.contour(X, Y, Z, zdir='y', offset=40, cmap='coolwarm')

ax.scatter(2, -2, 8, c='red', marker='o', s=10)
ax.scatter(0, 0, 0, c='red', marker='o', s=10)

# # Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)
# ax.set_ylabel('y')
# ax.set_xlabel('x')
# ax.set_zlabel('z')

plt.show()
fig
```
